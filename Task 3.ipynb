## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Reading  Data

```python
df=pd.read_csv("globalterrorismdb_0718dist.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df["Casualities"] = df["nwound"]+df["nkill"]
```

```python
df.shape
```

```python
print('Country with Highest Terrorist Attacks:',df['country_txt'].value_counts().index[0])
print('Maximum people killed in an attack:',df['nkill'].max())
print("Year with the most attacks:",df['iyear'].value_counts().idxmax())
print("Year with the least attacks:",df['iyear'].value_counts().idxmin())
print("Most Attack Types:",df['attacktype1_txt'].value_counts().idxmax())
print("Most Active Group:",df['gname'].value_counts().index[1])
```

## Static Visualization

```python
plt.figure(figsize=(20,7))
x=df["country_txt"].value_counts().index
sns.countplot(df['country_txt'], order=x[:20],palette=sns.color_palette("flare"))
plt.title("Top 20 Countries Under Threat")
plt.xlabel("Country")
plt.ylabel("Total Death")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(15,7))
y=df["attacktype1_txt"].value_counts().index
sns.countplot(df["attacktype1_txt"], order=y, palette = sns.color_palette("flare"))
plt.title("Forms Of Attack")
plt.xlabel("Weapons Used")
plt.ylabel("Total Death ")
plt.xticks(rotation=90)
```

```python
pd.crosstab(df.iyear,df.region_txt).plot(kind="area",figsize=(15,7))
plt.title('Terrorist Activities By Region In Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
```
